```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

read_file = pd.read_csv('gapminder-FiveYearData.txt')
read_file.to_csv('gapminder-FiveYearData.csv',index=None)
```

```python
df = pd.read_csv('gapminder-FiveYearData.csv')
df.head()
```

```python
#create a pivot table
table = pd.pivot_table(df,values='lifeExp',index=['continent'],columns=['year'])
table
```

```python
fig,ax = plt.subplots(figsize=(15,10))
ax = sns.heatmap(table,annot =True, center=0)
fig.savefig('SeabornPythonProgram.png')
```
